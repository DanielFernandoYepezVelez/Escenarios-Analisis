```python
""" Valores equivalentes a nulos (NaT, None, np.nan):
.notnull() funciona con cualquiera de estos valores, ya que son considerados "nulos" por Pandas. """
# Asignar A Una Nueva Variable
df_release = df_pivot[df_pivot['fecha_lanzamiento'].notnull()]
```

```python
df_release['año_lanzamiento'] = df_release['fecha_lanzamiento'].dt.year
```

```python
df_release.loc[:,'año_lanzamiento'] = df_release['fecha_lanzamiento'].dt.year
```

```python
df_release.info()
```

```python
df_release.head()
```

```python
movies_per_year = df_release.groupby(by='año_lanzamiento').size()
print(movies_per_year)
```

```python
# Asegurarse de que los años estén ordenados
movies_per_year.sort_index(inplace=True)

# Crear el rango completo de años
min_year = movies_per_year.index.min()
max_year = movies_per_year.index.max()
all_years = range(min_year, max_year + 1)

# Completar años faltantes con ceros
movies_per_year = movies_per_year.reindex(all_years, fill_value=0)

# Crear la figura y el gráfico
plt.figure(figsize=(12, 6))
plt.bar(movies_per_year.index, movies_per_year.values, color='skyblue', alpha=0.8)

# Personalizar el gráfico
plt.title('Número de Películas Lanzadas por Año', fontsize=16)
plt.xlabel('Año de Lanzamiento', fontsize=14)
plt.ylabel('Cantidad de Películas', fontsize=14)
plt.xticks(ticks=range(min_year, max_year + 1, 10), rotation=45)  # Mostrar etiquetas cada 10 años
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
