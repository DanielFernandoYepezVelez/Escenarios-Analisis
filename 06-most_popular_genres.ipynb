```python
# Paso 3: Separar géneros en filas
df_pivot = df_pivot.explode(column='generos')
```

```python
df_pivot = df_pivot[~(df_pivot['generos'].str.strip().eq(''))]
# df_pivot = df_pivot[(df_pivot['generos'].str.strip().str.len() > 0)]
```

```python
# Retorna Una Serie De Generos Y Sus Apraciones (Repetidos)
df_pivot['generos'].value_counts()
```

```python
appearances_genres = df_pivot['generos'].value_counts()
```

```python
# Agrupación A Nivel De Columna
appearances_genres.groupby('generos').sum()
```

```python
# Agrupación De Generos Por Valor O Dato
grouped_genres = appearances_genres.groupby(lambda x: x.strip()).sum()
```

```python
# Genero Más Popular Y Máximo De Aparaciones
most_popular_genre = grouped_genres.idxmax()
count_popular_genre = grouped_genres.max()

print(f"El género más popular es: {most_popular_genre} con {count_popular_genre} apariciones.")
```

```python
# Incrementar el tamaño del gráfico
plt.figure(figsize=(16, 8))

# Gráfico de barras
grouped_genres.plot(kind='bar', color='skyblue', alpha=0.8)

# Personalización
plt.title('Frecuencia de Géneros en Películas', fontsize=16)
plt.xlabel('Géneros', fontsize=14)
plt.ylabel('Cantidad de Películas', fontsize=14)
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar diseño
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
# Genero Menos Popular Y Menos De Aparaciones
less_popular_genre = grouped_genres.idxmin()
count_popular_genre = grouped_genres.min()

print(f"El género menos popular es: {less_popular_genre} con {count_popular_genre} aparición.")
```

```python
# Paso 2: Graficar
plt.figure(figsize=(8, 6))
plt.bar(less_popular_genre, count_popular_genre, color='salmon')

# Personalizar el gráfico
plt.title(f"Género Menos Popular: {less_popular_genre}", fontsize=14)
plt.xlabel("Género", fontsize=12)
plt.ylabel("Número de Películas", fontsize=12)

# Mostrar el gráfico
plt.show()
```
